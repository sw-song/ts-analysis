```python
import numpy as np
import pandas as pd
pd.options.display.float_format = '{:.3f}'.format
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import warnings
warnings.filterwarnings(action='ignore')
```

```python
x = np.array([np.round(x, 3) for x in np.arange(-0.99, 1, 0.01)])
y = np.log(1+x)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
axes[0].scatter(x=x,y=y, s=10, alpha=0.5, label='y=ln(1+x)')
axes[0].scatter(x=x,y=x, s=5, alpha=0.5, label='y=x')
axes[0].legend(labelcolor='w', facecolor='gray')
axes[0].fill_between(x=x, y1=x, y2=y, alpha=0.2)

axes[1].scatter(x=x,y=y, s=10, alpha=0.8, label='y=ln(1+x)')
axes[1].scatter(x=x,y=x, s=5, alpha=0.8, label='y=x')
axes[1].legend(labelcolor='w', facecolor='gray')
axes[1].set_xlim(-0.3, +0.3)
axes[1].set_ylim(-0.3, +0.3)
axes[1].fill_between(x=x, y1=x, y2=y, alpha=0.2)

plt.show()
```

```python
arg_idx = ((x>-0.1) & (x<0.1))
arg_idx
```

```python
import seaborn as sns
```

```python
df = pd.DataFrame(data={
        'x': x[arg_idx], 
        'y': y[arg_idx],
        '|x-y|': abs(x[arg_idx] - y[arg_idx]),
        'x(%)': x[arg_idx],
        'y(%)': y[arg_idx],
        })
         
to_pct = lambda x : f'{int(x*100)}%'
df['x(%)'] = df['x(%)'].apply(lambda x : to_pct(x))   
df['y(%)'] = df['y(%)'].apply(lambda x : to_pct(x))   

print(df)
sns.heatmap(df[['|x-y|']], annot=True, cmap='Blues_r')
plt.show()
```
