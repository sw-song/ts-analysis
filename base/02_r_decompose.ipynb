```r
library("forecast")
library("ggplot2")
```

```r
data(AirPassengers)
ap = AirPassengers
ap
```

```r
class(ap)
```

```r
options(repr.plot.width=10, repr.plot.height=6)
par(bg="white")
plot(ap)
```

```r
options(repr.plot.width=10, repr.plot.height=6)
par(bg="white")
plot(decompose(ap))
```

```r
options(repr.plot.width=10, repr.plot.height=6)
par(bg="white")
plot(decompose(ap, type="multi"))
```

```r
ap.decom = decompose(ap, type="multi")
```

```r
options(repr.plot.width=10, repr.plot.height=6)
par(bg="white")
plot(ap.decom$trend)
```

```r
options(repr.plot.width=10, repr.plot.height=6)
par(bg="white")
plot(ap.decom$season)
```

```r
trend = ap.decom$trend
season = ap.decom$season
```

```r
trend_with_season = cbind(trend, trend*season)
trend_with_season
```
