### AR(1)

- |Pi| < 1  (약정상성)
- |Pi| = 0  (단위근, 비정상성)
- |Pi| > 0  (무한대 발산, 비정상성)

```r
y = arima.sim(n=100, list(ar=0.9), innov=rnorm(100))
par(bg='white')
ggtsdisplay(y)
```

```r
y = arima.sim(n=100, list(ar=0.5), innov=rnorm(100))
par(bg='white')
ggtsdisplay(y)
```

```r
y = arima.sim(n=100, list(ar=0.1), innov=rnorm(100))
par(bg='white')
ggtsdisplay(y)
```

```r
y = arima.sim(n=100, list(ar=-0.1), innov=rnorm(100))
par(bg='white')
ggtsdisplay(y)
```

```r
y = arima.sim(n=100, list(ar=-0.5), innov=rnorm(100))
par(bg='white')
ggtsdisplay(y)
```

```r
y = arima.sim(n=100, list(ar=-0.9), innov=rnorm(100))
par(bg='white')
ggtsdisplay(y)
```
